```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
```

```python
data = pd.read_csv('../data/Train_call.txt',delimiter = '\t')
labels = pd.read_csv('../data/Train_clinical.txt',delimiter = '\t')
```

```python
data.head()
```

```python
data = data.transpose()
data
```

```python
data = data[4:]
data
```

```python
data = data.reset_index()
data
```

```python
data.rename(columns={'index':'Sample'}, inplace=True)
data.head()
```

```python
labels.head()
```

```python
# Combine labels and instances
combined = pd.merge(data,labels,on='Sample',)
combined
```

```python
X = combined.iloc[:,1:2835]
y = combined['Subgroup']

combined.head()
```

```python
# check if class lables are balanced or not
y.value_counts()
```

## Logistic regression

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.75, random_state=4)
print('X_train', X_train.shape)
print('X_test', X_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
```

```python
# without cross validation
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# build the model
lr_clf = LogisticRegression(random_state=0)

# load input to the model
lr_clf.fit(X_train,y_train)

# check the prediction score
lr_clf.score(X_test,y_test)
```
